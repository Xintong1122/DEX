# Aave

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create dataframe from data
data = pd.read_csv("https://raw.githubusercontent.com/Xintong1122/DEX/main/Data/Query_data/Aave_Queried.csv")
df = pd.DataFrame(data)

# convert date column to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
# split data before and after 2022-11-11
before_df = df[df['date'] < '2022-11-11']
after_df = df[df['date'] >= '2022-11-11']
```

```python
# Draw the broken line statistical graph
plt.plot(df['date'], df['val'], label='Decentralization Index')

# Draw the fitting line before 2022.11.11
before_fit = np.polyfit(before_df.index, before_df['val'], 1)
before_line = np.poly1d(before_fit)
plt.plot(before_df['date'], before_line(before_df.index), label='Before')

# Draw the fitting line after 2022.11.11
after_fit = np.polyfit(after_df.index, after_df['val'], 1)
after_line = np.poly1d(after_fit)
plt.plot(after_df['date'], after_line(after_df.index), label='After')

plt.axvline(x=pd.to_datetime('2022-11-11'), color='red', linestyle='-', label='FTX Event')

# Add legend and labels
plt.legend()
plt.xlabel('Date')
plt.xticks(fontsize=8)
plt.xticks(rotation=45)
plt.ylabel('Value of transaction decentralization')
plt.title('Aave')


# Show the plot
plt.show()
```
